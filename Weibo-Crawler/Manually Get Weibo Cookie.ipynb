#### Manually Get Weibo Cookie

__Step 1__<br>
Go to weibo.com.

__Step 2__<br>
Open Inspect mode. <br>
`1` Ctrl + Shift + C, or <br>
`2` Right-click any part of the page and choose Inspect Element.


__Step 3__ <br>
Go to network


__Step 4__<br>
Feel free to refresh the page and find `newlogin` from the sidebar. You can find your cookie here.
